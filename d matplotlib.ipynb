```python
#lets say i want compare them and see them side by side 

plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
##label and putting title

plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.plot(batsman['index'],batsman['V Kohli'],color='#345ceb')
plt.title('Rohit sharma va virat kohli career stats')
plt.xlabel('Season')
plt.ylabel('Runs Scored')


```

```python
#lets say we want dashed lines


plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed')
plt.plot(batsman['index'],batsman['V Kohli'],color='#345ceb', linestyle='dashdot')
plt.title('Rohit sharma va virat kohli career stats')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#increase the line width

plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linewidth=10)
plt.plot(batsman['index'],batsman['V Kohli'],color='#345ceb')
plt.title('Rohit sharma va virat kohli career stats')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#Marker : A marker is a symbol used to represent each point on a plot

plt.plot(batsman['index'],batsman['RG Sharma'],color='green', marker='D',markerfacecolor='yellow')
plt.plot(batsman['index'],batsman['V Kohli'],color='#345ceb',marker='*', markersize=10,markerfacecolor='red')
plt.title('Rohit sharma va virat kohli career stats')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#legends
# which line represents which player

plt.plot(batsman['index'],batsman['RG Sharma'],color='green', marker='D',markerfacecolor='yellow',label='ViratK')
plt.plot(batsman['index'],batsman['V Kohli'],color='#345ceb',marker='*', markersize=10,markerfacecolor='red',label='RohitS')
plt.legend()
plt.grid()
plt.title('Rohit sharma va virat kohli career stats')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
df = pd.read_csv('batter.csv')
```

```python
df
```

```python
plt.scatter(df['avg'], df['strike_rate'])
plt.title('Average and strike rate of top batsman')
plt.xlabel('average')
plt.ylabel('Strikerate')
```

```python
# i want to find the top 50 players's strike rate(by runs)  


top50 = df.sort_values(by='runs',ascending=False).head(50)
# sort_values -> sort the data
# ascending= False -> higest first
#head(50) -> top 50 rows

top50.head()

plt.scatter(top50['avg'], top50['strike_rate'])
plt.title('Average and strike rate of top batsman')
plt.xlabel('average')
plt.ylabel('Strikerate')
```

```python
#show name of the player as well of top 5 players

top5 = df.sort_values(by='runs',ascending=False).head()
# sort_values -> sort the data
# ascending= False -> higest first
#head(50) -> top 50 rows

top5.head()

plt.scatter(top5['avg'], top5['strike_rate'])

for i in range(len(top5)):
    plt.annotate(top5['batter'].iloc[i],(top5['avg'].iloc[i],top5['strike_rate'].iloc[i]),xytext=(5,5),textcoords='offset points')


plt.title('Average and strike rate of top 5 batsman')
plt.xlabel('average')
plt.ylabel('Strikerate')

```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'],color=['green','yellow','red','orange','pink'])
```

```python
# labels on the bars
plt.bar(df['batsman'],df['2015'],color=['green','yellow','red','orange','pink'])
for i in range(len(df)):
    plt.text(i, df['2015'].iloc[i], str(df['2015'].iloc[i]),ha='center',va='bottom') #ha - horizontal alignment #va - vertical alignment
    #Plt.txt means put some text at (x,y) position on ur graph
    
plt.show()
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
```

```python
## there is no comparison as the graph is overlapping 
# we can take a solution 
y = df.shape[0] #show the number of rows

x = np.arange(df.shape[0]) ## give me array -[0,1,2,3,4]
plt.bar(x-0.2,df['2015'],width=0.2,color="yellow")
plt.bar(x,df['2016'],width=0.2,color="green")
plt.bar(x+0.2,df['2017'],width=0.2,color="pink")
plt.xticks(x,df['batsman'])
plt.show()


```
