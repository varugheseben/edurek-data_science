```python
# Define a list of 10 sentences which has positive and negative sentimence
# Label the sentecene with +ve or -ve sentimence
# Tokenize the sentences
# Remove Stopwords
# Use Porter Stemmer and print stemmed tokens
# Do pos tagging
# Use RegEx parser to give grammer of NP(Noun Phrase)
# Ceate list of stemmed tokens and do TFIDF vectorization
# use Logistic Regression to create the model and train the model using data obtained via vectorization
# Take a new sentence and determine whether its a positive or negative sentimence

import nltk
from nltk import word_tokenize, pos_tag
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression

nltk.download('stopwords')
nltk.download('punkt_tab')

english_stop_words = stopwords.words('english')

def remove_stopwords(tokens):
  cleaned_tokens = [token for token in tokens if token.lower() not in english_stop_words]
  return cleaned_tokens

sentences = [
    "The laptop performs exceptionally well and has outstanding battery life.",
    "This smartphone has a stunning display and a very responsive interface.",
    "The headphones provide crystal-clear sound and are comfortable to wear all day.",
    "The coffee maker is easy to use and brews excellent coffee every time.",
    "The smartwatch offers useful features and excellent fitness tracking.",
	  "The vacuum cleaner stopped working after only a few weeks of use.",
    "This tablet is slow, frequently freezes, and has poor battery life.",
    "The gaming mouse feels cheap and its buttons often fail to register clicks.",
    "The refrigerator consumes too much electricity and makes excessive noise.",
    "The blender has a weak motor and struggles to blend even soft ingredients."
]

labels =[
    1, # Positive
    1, # Positive
    1, # Positive
    1, # Positive
    1, # Positive
    0, # Negative
    0, # Negative
    0, # Negative
    0, # Negative
    0 # Negative
]

cleaned_tokens_list = [ remove_stopwords(word_tokenize(sentence)) for sentence in sentences ]

stemmed = {
    'words': list(),
    'stem': list()
}
ps = PorterStemmer()
for cleaned_tokens in cleaned_tokens_list:
  for token in cleaned_tokens:
    stemmed['words'].append(token)
    stemmed['stem'].append(ps.stem(token))

pd.set_option('display.max_rows', None)
data_frame = pd.DataFrame(stemmed)

cleaned_tokens_list = [ " ".join(sentence) for sentence in cleaned_tokens_list ]
vectorizer = TfidfVectorizer()
# We can either use the sample data or cleaned data for vectorization
# X = vectorizer.fit_transform(sentences) # Using sample data for vectorization
X = vectorizer.fit_transform(cleaned_tokens_list) # Using cleaned data for vectorization
Y = labels

sentiment_analyze_model = LinearRegression()
sentiment_analyze_model.fit(X,Y)

sentence = input("Enter sentence to analyze: ")
sentence_X = vectorizer.transform([sentence])
print(sentence_X)
sentimence_predicted = sentiment_analyze_model.predict(sentence_X)
print(sentimence_predicted)

if (sentimence_predicted == 0.50):
  print("Neutral sentence")
elif (sentimence_predicted > 0.50):
  print("Positive Sentence")
else:
  print("Negative Sentence")
```
