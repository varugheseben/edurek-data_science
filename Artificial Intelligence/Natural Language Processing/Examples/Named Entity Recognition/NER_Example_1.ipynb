```python
from nltk import word_tokenize, pos_tag, ne_chunk
text = "In 2026, Sundar Pichai announced that Google would expand its offices in Newark."

tokens = word_tokenize(text)
pos_tags = pos_tag(tokens)

ner_tree = ne_chunk(pos_tags)
ner_tree

# In the below tree, some of the entities are not recognized correctly
# For example, year 2026 is not classified as DATE
#              Google is not classified as ORGANIZATION
# This is due to old classification models doing the prediction wrongly.
# So we need to use advanced deep learning models like spaCy.
```

```python
# Lets use spaCy to do NER
import spacy
import pandas as pd

nlp = spacy.load('en_core_web_sm')
text = "In 2026, Bill Gates announced that Microsoft would expand its offices in Newark. Microsoft is looking at buying UK startup for $1 billion. Bill Gates founded Azure."
doc = nlp(text)

entities = [(ent.text, ent.label_) for ent in doc.ents]

data_frame = pd.DataFrame(entities, columns=['Text', 'Entity Recognized'])
data_frame
```
