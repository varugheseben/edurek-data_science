```python
# Bydefault separator used is dot character
import nltk
from nltk.tokenize import sent_tokenize

prompt = input('Enter you sentences: ')
tokens = sent_tokenize(prompt)
print(tokens)
```

```python
# If wanted to use separate character instead of dot then we need to use RegexpTokenizer
# In below example we are using ; as separator
import nltk
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'[^;]+')
prompt = input('Enter you sentences: ')
tokens = tokenizer.tokenize(prompt)
print(tokens)
```
