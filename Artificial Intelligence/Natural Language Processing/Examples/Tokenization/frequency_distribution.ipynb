```python
# Frequency distribution is used for determining repeatness of each token
# In the case of word tokenization

import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

prompt = input("Enter your query: ")
tokens = word_tokenize(prompt)
frequency_distribution = FreqDist(tokens)

for word, frequency in frequency_distribution.items():
  print(f'{word} is repeated {frequency} times')

frequency_distribution.plot(10)
```

```python
# Frequency distribution for sentence tokenization
import nltk
from nltk.tokenize import sent_tokenize
from nltk.probability import FreqDist

prompt = input("Enter your query: ")
tokens = sent_tokenize(prompt)
frequency_distribution = FreqDist(tokens)

for sentence, frequency in frequency_distribution.items():
  print(f'{sentence} is repeated {frequency} times')

```
