```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

text = "Python has many libraries which can be used for building systems with artificial intelligence"
tokens = word_tokenize(text)
ngrams_list = ngrams(tokens, 1)

print(f'N-grams of {text}')
for ngram in (list(ngrams_list)):
  print(ngram)
```
