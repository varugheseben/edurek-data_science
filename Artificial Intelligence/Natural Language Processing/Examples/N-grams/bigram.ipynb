```python
import nltk
from nltk.util import bigrams
from nltk.tokenize import word_tokenize

text = "Python has many libraries which can be used for building systems with artificial intelligence"
tokens = word_tokenize(text)
bigrams_list = bigrams(tokens)

print(f'Bigrams of {text}')
for bigram in (list(bigrams_list)):
  print(bigram)
```
