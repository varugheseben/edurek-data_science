```python
import nltk
from nltk.util import trigrams
from nltk.tokenize import word_tokenize

text = "Python has many libraries which can be used for building systems with artificial intelligence"
tokens = word_tokenize(text)
trigrams_list = trigrams(tokens)

print(f'Trigrams of {text}')
for trigram in (list(trigrams_list)):
  print(trigram)
```
