```python
# Read a text file
# tokenize the text
# remove the stop words
# find pos tags
# apply stemming
# apply lemmatization
# The output should have unique words without punctuations line .,()\n\t etc.
# output format is
# word      pos meaning stem      lemma
# children  NNS Noun    children  child

import nltk
nltk.download('omw-1.4') # This is used for fetching dictionary of words along wioth the meanings
nltk.download('averaged_perceptron_tagger') # used to find part of speech tagging
nltk.download('averaged_perceptron_tagger_eng') # used for lemmatization
nltk.download('punkt') # used for tokenizartion
nltk.download('wordnet') # used for lemmatizations, synonyms, antnonyms
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
from operator import pos
from nltk.stem import SnowballStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords,wordnet
from nltk import pos_tag
import pandas as pd

def read_file(file_name):
  contents = list()
  with open(file_name, "r") as file:
    for line in file:
      contents.append(line.strip())

  return contents

def remove_stop_words(tokens, additional_set=None):
  stop_words = set(stopwords.words('english'))
  if additional_set is not None:
    stop_words.update(additional_set)

  return [token for token in tokens if token.lower() not in stop_words]

def get_pos_tag(tag):
  if tag.startswith('J'):
    return wordnet.ADJ
  elif tag.startswith('V'):
    return wordnet.VERB
  elif tag.startswith('N'):
    return wordnet.NOUN
  elif tag.startswith('R'):
    return wordnet.ADV
  else:
    return wordnet.NOUN

def tag_meaning(tag):
  if tag.startswith('J'):
    return 'Adjective'
  elif tag.startswith('V'):
    return 'Verb'
  elif tag.startswith('N'):
    return 'Noun'
  elif tag.startswith('R'):
    return 'Adverb'
  else:
    return 'Noun'

contents = read_file('sample_data/data.txt')
contents_stage_1 = list(set(contents)) # Cleaning to remove duplicates from content

lemmatizer = WordNetLemmatizer()
sb = SnowballStemmer('english')
data_set = dict()

for index in range(len(contents)):
  tokens = list(set(word_tokenize(contents[index])))
  cleaned_tokens = remove_stop_words(tokens) # Removing the stop words
  pos_tagged = pos_tag(cleaned_tokens)
  for word, tag in pos_tagged:
    if not word in data_set:
      data_set[word] = {
          'Line Number': (index + 1),
          'Meaning': tag_meaning(tag),
          'Stem': sb.stem(word),
          'Lemma': lemmatizer.lemmatize(word, pos=get_pos_tag(tag)),
      }
    else:
      data_set[word]['Line Number'] = "{},{}".format(data_set[word]['Line Number'], (index + 1))

pd.set_option('display.max_columns', None) # To display all the columns of data frame instead of truncating
data_frame = pd.DataFrame(data_set)
data_frame
```
