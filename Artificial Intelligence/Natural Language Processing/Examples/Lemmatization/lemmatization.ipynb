```python
from nltk.stem import WordNetLemmatizer

words = ['running', 'better', 'studies', 'went']

lemmatizer = WordNetLemmatizer()

pos_values = {
    'v': 'Verb',
    'n': 'Noun',
    'a': 'Adjective',
}

for word in words:
  for pos_value in pos_values.keys():
    print(f'Word[{word}], POS[{pos_values[pos_value]}], Lemma[{lemmatizer.lemmatize(word, pos=pos_value)}]')
  print('-----------------------------------------------------')
```
