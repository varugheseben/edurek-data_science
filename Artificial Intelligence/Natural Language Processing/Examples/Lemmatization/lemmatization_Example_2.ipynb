```python
# Take a sentence from the user as input
# Tokenize the whole sentence into individual tokens
# Run a for loop to print original word with its lemmatized word

import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('all')
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

sentence = input("Enter your sentence: ")
lemmatizer = WordNetLemmatizer()

pos_values = {
    'v': 'Verb',
    'a': 'Adjective',
    'n': 'Noun'
}

for token in word_tokenize(sentence):
  for pos_value in pos_values.keys():
    print(f'Word[{token}, POS[{pos_values[pos_value]}], Lemma[{lemmatizer.lemmatize(token, pos=pos_value)}]')
  print('-' * 45)
```
