```python
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

user_input = input("Enter a sentence: ")
tokens = word_tokenize(user_input)

stop_words = set(stopwords.words('english')) # Get all the English stop words
cleaned_tokens = [token for token in tokens if token.lower() not in stop_words]

print(f'Original Sentence: {user_input}')
print(f'Tokens before removing stopwords: {tokens}')
print(f'Cleaned tokens: {cleaned_tokens}')
# Note: Here full stop is not removed as a stop word. Because its the sentance separator
```
