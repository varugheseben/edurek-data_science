```python
# Example of LancasterStemmer
from nltk.stem import LancasterStemmer

ls = LancasterStemmer()

words = ['Running', 'Runs', 'Ran', 'Runner']

for w in words:
  print(w, ls.stem(w),sep=" -> ")

# Its not correctly finding stem for all words
```
