```python
# Take a input from the user and tokenize the sentence and use PorterStemmer, LancasterStemmer and SnowballStemmer to find the stem of the words

from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
from nltk.tokenize import word_tokenize
import pandas as pd

prompt = input("Enter your query: ")

tokens = word_tokenize(prompt)

ps = PorterStemmer()
ls = LancasterStemmer()
sb = SnowballStemmer('english')

stems = {
    'words': [],
    'porter_stemmer': [],
    'lancaster_stemmer': [],
    'snowball_stemmer': [],
}

for token in tokens:
  stems['words'].append(token)
  stems['porter_stemmer'].append(ps.stem(token))
  stems['lancaster_stemmer'].append(ls.stem(token))
  stems['snowball_stemmer'].append(sb.stem(token))

data_frame = pd.DataFrame(stems)
data_frame
```
