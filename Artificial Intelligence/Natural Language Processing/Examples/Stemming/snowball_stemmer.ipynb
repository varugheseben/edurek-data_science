```python
# Example of Snowball Stemmer
from nltk.stem import SnowballStemmer

sb = SnowballStemmer('english')

words = ['Running', 'Runs', 'Ran', 'Runner']

for w in words:
  print(w, sb.stem(w),sep=" -> ")

#Its not generating correct stems for all words
```
