```python
#Example of Porter Stemmer
from nltk.stem import PorterStemmer
ps = PorterStemmer()

words = ['Running', 'Runs', 'Ran', 'Runner']

for w in words:
  print(w,ps.stem(w),sep=" -> ")

# Its not correctly finding stem for all words
```
