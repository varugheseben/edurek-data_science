```python
import pandas as pd

documents = [
    "I love machine learning",
    "Machine learning is fun",
    "I love coding"
]

tokens = [ doc.lower().split() for doc in documents]

vocab = sorted(set(word for doc in tokens for word in doc))

bow_vectors = []

for doc in tokens:
  word_count = dict.fromkeys(vocab, 0)
  for word in doc:
    word_count[word] += 1

  bow_vectors.append(word_count)

df = pd.DataFrame(bow_vectors)
df
```
