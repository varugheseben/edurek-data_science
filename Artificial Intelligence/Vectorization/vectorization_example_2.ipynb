```python
# This is example of count vectorization without python library

import nltk
import pandas as pd
from nltk import word_tokenize

nltk.download('punkt_tab')

sentences = [
    "The laptop performs exceptionally well and has outstanding battery life.",
    "This smartphone has a stunning display and a very responsive interface.",
    "The headphones provide crystal-clear sound and are comfortable to wear all day.",
    "The coffee maker is easy to use and brews excellent coffee every time.",
    "The smartwatch offers useful features and excellent fitness tracking.",
	  "The vacuum cleaner stopped working after only a few weeks of use.",
    "This tablet is slow, frequently freezes, and has poor battery life.",
    "The gaming mouse feels cheap and its buttons often fail to register clicks.",
    "The refrigerator consumes too much electricity and makes excessive noise.",
    "The blender has a weak motor and struggles to blend even soft ingredients."
]

uniq_vocabulary = set()

for sentence in sentences: uniq_vocabulary.update(word_tokenize(sentence))

data_set = []
for sentence in sentences:
  row = [1 if word in word_tokenize(sentence) else 0 for word in uniq_vocabulary]
  data_set.append(row)

data_frame = pd.DataFrame(data_set, columns=list(uniq_vocabulary))
data_frame
```
