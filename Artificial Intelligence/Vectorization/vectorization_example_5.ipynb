```python
import pandas as pd
documents = [
    "This is first senetence",
    "This is second sentence",
    "This is third sentence"
]

vocab = {}
for sentence in documents:
  for token in sentence.split():
    vocab[token] = 1

df = pd.DataFrame([{col:0 for col in vocab } for i in range(len(documents))])

for i in range(len(documents)):
  for token in documents[i].split():
    if token in vocab:
      df.at[i, token] = 1

df
```
