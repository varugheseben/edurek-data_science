```python
# TFIDF Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

sentences = [
    "The cat is sitting on a mat and cat is drinking milk.",
    "This smartphone has a stunning display and stunning look.",
    "The headphones provide crystal-clear sound and also provide smooth sound.",
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)
print(vectorizer.get_feature_names_out())
print(X.toarray())

pd.set_option('display.max_columns', None)
data_frame = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
data_frame
```
