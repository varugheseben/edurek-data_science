```python
import numpy as np
from scipy import stats

salaries = [45000, 50000, 50000, 53000, 55000, 200000]

print('Mean:  ', np.mean(salaries))      # Pulled up by 200k
print('Median:', np.median(salaries))    # Robust middle
print('Mode:  ', stats.mode(salaries).mode)  # Most common
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Two classes, same average score
class_a = [70, 72, 68, 71, 69]   # Low spread
class_b = [50, 90, 40, 95, 75]   # High spread
# class a -> values are close -> low variation
# class b -> values are far apart -> high deviation
print(f"class a = Mean: {np.mean(class_a)}")
print(f"class b = Mean: {np.mean(class_b)}")
print (f"class A deviation is {round(np.std(class_a), 2)}")  # rounding to two decimal values
print (f"class B deviation is {round(np.std(class_b), 2)}")


# print('Class A - Mean:', np.mean(class_a), '| SD:', round(np.std(class_a), 2))
# print('Class B - Mean:', np.mean(class_b), '| SD:', round(np.std(class_b), 2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
#given data
data = [0,10,20,30,40,50,60,70,80,190,210]
#quatiles
Q1 = np.percentile(data,25)
Q2 = np.percentile(data,50)
Q3 = np.percentile(data,75)
print("Q1(25%) : ", Q1)
print("Q2(Median->50%):  ",Q2)
print("Q3(75%)" ,Q3)
plt.boxplot(data)
plt.ylabel("values")
plt.show()

```

```python
# dice roll simulation ..probabilty of even number

import numpy as np
#simulate 10,000 dice rolls
rolls = np.random.randint(1,7,10000)

#print(rolls, rolls % 2 == 0)
# probility of even number
prob = np.sum(rolls % 2 == 0) / len(rolls)
print(f"Probabilty of even is : {prob}") 
```
