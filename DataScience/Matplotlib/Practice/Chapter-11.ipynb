```python
# Removing the outlier.
# An outlier is a data point that is significantly different from the other observations in a dataset.
# Think of it as the "odd one out"—a value that lies an abnormal distance from other values in a random sample from a population.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

price = np.array([10000, 12000, 13000, 14000, 13000, 1200000000])
year = np.array([2015, 2016, 2017, 2018, 2019, 2020])
plt.plot(year, price)
plt.show()
# In the below graph we have an outlier at 2020 so the graph before that one show as a straight line.
# So it will be difficult to anlayze the trend during the year 2015 to 2019.
# So the solution what we can do is remove the outlier from the graph
```

```python
# We can remove outliers from both axis, for that we need to use xlim() and ylim()
# xlim() for removing the outlier from X-axis
# ylim() for removing the outlier from Y-axis
# So from above graph straight line is showing for data set from 2015 to 2019. So will limit the dataset for that period
plt.xlim(2015, 2019)
plt.ylim(9000, 20000)
plt.plot(year, price)
plt.show()
```

```python
# In the above graph issue on the values of ticks. So we need to specify to use ticks from 'price' array.
# Remember we removed outliers from x-axis so we cannot use all values of price. We need to slice that and use only relevant to 
# trimmed x-axis values. That means we need to remove 2020 from x-axis
plt.xlim(2015, 2019)
plt.ylim(9000, 20000)
plt.xticks(year[:5])
plt.plot(year, price)
plt.show()
```
