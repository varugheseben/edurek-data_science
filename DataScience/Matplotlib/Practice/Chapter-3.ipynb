```python
# Adding title to the grpah
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y = np.array([15, 35, 56, 25, 45, 40, 35])

graph_style = dict(marker='*', 
         markersize=10, 
         markerfacecolor='blue',
         markeredgecolor='black',
         linestyle='solid', # supported styles are dashed, dotted,dashdot,None
         linewidth=4,
         color='red')

plt.plot(x,y, **graph_style)
plt.title('Class Size', 
          fontsize=20, 
          family='DejaVu Sans',
          fontweight='bold',
          color='blue'
         )
plt.show()
```

```python
# Adding X and Y label to the grpah
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y = np.array([15, 35, 56, 25, 45, 40, 35])

graph_style = dict(marker='*', 
         markersize=10, 
         markerfacecolor='blue',
         markeredgecolor='black',
         linestyle='solid', # supported styles are dashed, dotted,dashdot,None
         linewidth=4,
         color='red')

plt.plot(x,y, **graph_style)
plt.xlabel('Year',
          fontsize=20, 
          family='DejaVu Sans',
          fontweight='bold',
          color='green'
          )
plt.ylabel('Students',
          fontsize=20, 
          family='DejaVu Sans',
          fontweight='bold',
          color='green'
          )

plt.show()
```

```python
# Define the X and Y axis ticks similar to coordinates
# Adding X and Y label to the grpah
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y = np.array([25, 45, 40, 35])

graph_style = dict(marker='*', 
         markersize=10, 
         markerfacecolor='blue',
         markeredgecolor='black',
         linestyle='solid', # supported styles are dashed, dotted,dashdot,None
         linewidth=4,
         color='red')

plt.plot(x,y, **graph_style)
plt.xlabel('Year',
          fontsize=20, 
          family='DejaVu Sans',
          fontweight='bold',
          color='green'
          )
plt.ylabel('Students',
          fontsize=20, 
          family='DejaVu Sans',
          fontweight='bold',
          color='green'
          )

plt.xticks(x) # This will use values of x as the different tickts on x axis
plt.yticks(y) # This will use values of y as the different tickts on y axis
plt.show()
```

```python
# Customising different value points in X and Y axis
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y = np.array([25, 45, 40, 35])

graph_style = dict(marker='*', 
         markersize=10, 
         markerfacecolor='white',
         markeredgecolor='black',
         linestyle='solid', # supported styles are dashed, dotted,dashdot,None
         linewidth=4,
         color='red')

plt.plot(x,y, **graph_style)
plt.xlabel('Year',
          fontsize=20, 
          family='DejaVu Sans',
          fontweight='bold',
          color='green'
          )
plt.ylabel('Students',
          fontsize=20, 
          family='DejaVu Sans',
          fontweight='bold',
          color='green'
          )

plt.tick_params(axis='both', colors='blue')
plt.xticks(x) # This will use values of x as the different tickts on x axis
plt.yticks(y) # This will use values of y as the different tickts on y axis
plt.show()
```
