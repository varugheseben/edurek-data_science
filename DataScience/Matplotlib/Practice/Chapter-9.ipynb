```python
# Creating subplots. With subplot we can create multiple plots in a single canvas.
# So entire canvas is called Figure and each subplot in the canvas is called axes.
# To create subplots we can use subplots(). 
# subplots() returns a tuple and first element of tuple will be Figure object and second element will be array of axes
import matplotlib.pyplot as plt
import numpy as np

# So here we have first element as Figure object and second element as array of axes containing 4 axes.
# That's because we passed 2,2 to subplot(). So 2 rows and 2 columns
# We can have different value of (rows X columns) based on number of subplot we need
print(plt.subplots(2, 2))
```

```python
# Now instead of printing the return value of subplots() we will unpack its return value
figure, axes = plt.subplots(2,2)
x = np.array([1,2,3,4,5,6,7])

# Now we can plot the graph on each subplot
# We dont necessarily use plot() everytime. We can use other chart functions like bar(), pie()..
axes[0, 0].plot(x, x*2, color='blue')
axes[0, 0].set_title('x*2')

axes[0, 1].plot(x, x**2, color='yellow')
axes[0, 1].set_title('x**2')

axes[1, 0].plot(x, x**3, color='red')
axes[1, 0].set_title('x**3')

axes[1, 1].plot(x, x**4, color='green')
axes[1, 1].set_title('x**4')

# tight_layout() is used for not to overlap the labels and markers of the subplots
plt.tight_layout()
plt.show()
```

```python
figure, axes = plt.subplots(2,2)
x = np.array([1,2,3,4,5,6,7])

# Now we can plot the graph on each subplot
# We dont necessarily use plot() everytime. We can use other chart functions like bar(), pie()..
axes[0, 0].plot(x, x*2)
axes[0, 0].set_title('x*2')

axes[0, 1].bar(x, x**2)
axes[0, 1].set_title('x**2')

axes[1, 0].scatter(x, x**3)
axes[1, 0].set_title('x**3')

axes[1, 1].plot(x, x**4)
axes[1, 1].set_title('x**4')

# tight_layout() is used for not to overlap the labels and markers of the subplots
plt.tight_layout()
plt.show()
```
