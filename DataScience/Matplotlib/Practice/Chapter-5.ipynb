```python
# Creating bar chart using bar(). This will create vertical bars
import matplotlib.pyplot as plt
import numpy as np

categories = np.array(["Grains", "Fruits", "Vegetables", "Protein", "Dairy", "Sweets"])
quantity = np.array([4, 3, 2, 5, 3, 1])

plt.bar(categories, quantity, color="green")
plt.xlabel('Food')
plt.ylabel('Quantity')
plt.show()
```

```python
# Creating bar chart using barh(). This will create horizontal bars
import matplotlib.pyplot as plt
import numpy as np

categories = np.array(["Grains", "Fruits", "Vegetables", "Protein", "Dairy", "Sweets"])
quantity = np.array([4, 3, 2, 5, 3, 1])

plt.barh(categories, quantity, color="green")
plt.xlabel('Quantity')
plt.ylabel('Food')
plt.show()
```
