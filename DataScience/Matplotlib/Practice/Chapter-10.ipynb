```python
# Matplotlib + Panda
import matplotlib.pyplot as plt
import pandas as pd

pokemons = pd.read_csv('data/pokemon.csv')
# We going to get the value of count of pokemons based on Type1. That means we will get total count by Type1
type_count = pokemons['Type1'].value_counts(ascending=True)

plt.barh(type_count.index, type_count.values, color='red', edgecolor='black')
plt.title("No. of Pokemons by Type")
plt.xlabel('Count')
plt.ylabel('Type')
plt.tight_layout()
plt.show()
```
