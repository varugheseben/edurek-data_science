```python
# Plotting histogram using hist()
# Histogram is a visual representation of the distribution of quantitative data.
# It group values into bins(intervals) and counts how many falls in each range
import matplotlib.pyplot as plt
import numpy as np

# Lets say we going to create a numpy array of 100 scores with which median of the list will be 80 and different scores scale with difference of 10
# with median
scores = np.random.normal(loc=80, scale=10, size=100)

# In below data set most of the values are clustered near to 80. So if we increase the value of scale then values scattering will be more away from 80
scores
```

```python
# We also need to make sure values are not less than 0 and greater than 100
# For that purpose we can use clip()
scores = np.clip(scores, 0, 100) # This means value < 0 will become 0 and value > 100 will become 100

# hist() can be used for generating the histogram.
# We can pass additional parameters to hist()
# color -> For changing color of the histogram bins
# bins -> Number of bins
# edgecolor -> Edge color for each bin
plt.hist(scores, color='blue', bins=10, edgecolor='black')

plt.title('Test Scores')
plt.xlabel('Marks')
plt.ylabel('No. of Students')
plt.show()
```
