```python
# grid() for displaying grids in graph
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 10, 15, 20, 25])

plt.grid() # If no arguments passed then grid will show for both x and y axis
plt.plot(x,y)
plt.show()
```

```python
# grid() for displaying grids in graph
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 10, 15, 20, 25])

# We can pass additional parameters to grid()
# linewidth to increase the width of the line
# color to set the color for grid line
# linestyle for setting style of the grid line. Supported values are solid,dashed,dotted,dashdot
plt.grid(axis='both', linewidth=2, color='black', linestyle='solid')
plt.plot(x,y)
plt.show()
```
