```python
# Scatter graph
import pandas as pd
import matplotlib.pyplot as plt

batters = pd.read_csv('data/batter.csv')

# Fetch top 50 batsman's score card
top50 = batters.sort_values(by='runs', ascending=False).head(50)
# Draw scatter graph for top50
plt.scatter(top50['avg'], top50['strike_rate'])
plt.title("Score card of Top 50 Batsman")
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.show()
```

```python
# Fetch top 5 batsman and show their names on graph points
top5 = batters.sort_values(by='runs', ascending=False).head()

plt.scatter(top5['avg'], top5['strike_rate'])

for i in range(len(top5)):
    plt.annotate(
        top5['batter'].iloc[i], # This will be text to display
        (top5['avg'].iloc[i], top5['strike_rate'].iloc[i]), # Will be (x,y) coordinates of the data point
        xytext=(5,5), # Displacement of text. (5,5) means 5pixel right and 5 pixel up
        textcoords='offset points' # This means text coordinates need to be determined relative to the data point
    )
plt.show()
```
