```python
# Customizing the plot
# Different customization we can do graphs are
#  1) Add Marker
#  2) Add Marker Size
#  3) Add Marker Face Color
#  4) Add Marker Edge Color
#  5) Add line style
#  6) Add line width
#  7) Add line color

import matplotlib.pyplot as plt
import numpy as np

x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y = np.array([15, 35, 56, 25, 45, 40, 35])

plt.plot(x,y, 
         marker='*', 
         markersize=10, 
         markerfacecolor='blue',
         markeredgecolor='black',
         linestyle='solid', # supported styles are dashed, dotted,dashdot,None
         linewidth=4,
         color='red'
        )
# For colors we can use RGB value or HEX color code
plt.show()
```

```python
# Plotting multiple lines on same graph
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y1 = np.array([15, 35, 56, 25, 45, 40, 35])
y2 = np.array([25, 45, 36, 15, 55, 20, 50])
y3 = np.array([5, 30, 26, 35, 75, 45, 85])

plt.plot(x,y1, 
         marker='*', 
         markersize=10, 
         markerfacecolor='blue',
         markeredgecolor='black',
         linestyle='solid', # supported styles are dashed, dotted,dashdot,None
         linewidth=4,
         color='red'
        )

plt.plot(x,y2, 
         marker='^', 
         markersize=10, 
         markerfacecolor='green',
         markeredgecolor='black',
         linestyle='dashed', # supported styles are dashed, dotted,dashdot,None
         linewidth=4,
         color='blue'
        )

plt.plot(x,y3, 
         marker='o', 
         markersize=10, 
         markerfacecolor='cyan',
         markeredgecolor='black',
         linestyle='dotted', # supported styles are dashed, dotted,dashdot,None
         linewidth=4,
         color='green'
        )

plt.show()
```

```python
# We can use dictionary to store graph properties

import matplotlib.pyplot as plt
import numpy as np

x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y = np.array([15, 35, 56, 25, 45, 40, 35])

graph_style = dict(marker='*', 
         markersize=10, 
         markerfacecolor='blue',
         markeredgecolor='black',
         linestyle='solid', # supported styles are dashed, dotted,dashdot,None
         linewidth=4,
         color='red')

plt.plot(x, y, **graph_style)
plt.show()
```

```python
# Another implementation
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y = np.array([15, 35, 56, 25, 45, 40, 35])

graph_style = dict(marker='*', 
         markersize=10, 
         markerfacecolor='red',
         markeredgecolor='black',
         linestyle='solid', # supported styles are dashed, dotted,dashdot,None
         linewidth=4)

plt.plot(x, y, color='green', **graph_style) # Here we are passing the 'color' separately
plt.show()
```
