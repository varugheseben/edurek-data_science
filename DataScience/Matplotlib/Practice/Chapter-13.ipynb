```python
# Now display each data point value on the graph
# For that we can user text()
# text(x, y, s, fontdict=None, **kwargs)
# Bar chart
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Lets say we have data of students belongs to different class divisions
no_of_students = np.array([10, 25, 15, 30, 45, 18])
class_divisions = np.array(['10A', '10B', '10C', '10D', '10E', '10F'])

plt.bar(class_divisions, no_of_students, color=['green', 'red', 'blue', 'grey', 'black', 'yellow'])
for i in range(len(no_of_students)):
    plt.text(i, no_of_students[i], str(no_of_students[i]), ha='center', va='bottom')
plt.show()
```

```python
# Now lets say we have multiple y values to use against x axis. In such case we need to plot graph using width.
# Other wise it will overlap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

season_records = pd.read_csv('data/batsman_season_record.csv')
plt.bar(season_records['batsman'], season_records['2015'])
plt.bar(season_records['batsman'], season_records['2016'])
plt.bar(season_records['batsman'], season_records['2017'])
plt.show()
# In below graph its overlapping. So we need to use width the bar so that each y axis value can be seen separately.
# Please see below for that solution
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

season_records = pd.read_csv('data/batsman_season_record.csv')
x_coords = np.arange(season_records.shape[0])
x_coords 
plt.bar(x_coords - 0.2, season_records['2015'], width=0.2, color='blue', label='2015')
plt.bar(x_coords, season_records['2016'], width=0.2, color='red', label='2016')
plt.bar(x_coords + 0.2, season_records['2017'], width=0.2, color='green', label='2017')
plt.legend()
plt.xticks(x_coords, season_records['batsman'])
plt.show()
```
