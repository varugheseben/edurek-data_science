```python
# Basic example using python list
import matplotlib.pyplot as plt

x = [2020, 2021, 2022, 2023, 2024, 2025, 2026]
y = [15, 35, 56, 25, 45, 40, 35]

plt.plot(x,y)
plt.show() # To display the graph we need to call show()
```

```python
# Basic example using Numpy array
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
y = np.array([15, 35, 56, 25, 45, 40, 35])

plt.plot(x,y)
plt.show() # To display the graph we need to call show()
```

```python
# If we omit x coordinates it will generate x coordinates automatically 
import matplotlib.pyplot as plt
import numpy as np

y = np.array([15, 35, 56, 25, 45, 40, 35])

plt.plot(y)
plt.show() # To display the graph we need to call show()
```
