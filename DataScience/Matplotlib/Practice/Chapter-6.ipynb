```python
# Creating pie chart using pie()
import matplotlib.pyplot as plt
import numpy as np

categories = ['Freshmen', 'Sophomores', 'Juniors', 'Seniors']
values = np.array([300, 150, 200, 225])
colors = ['red', 'green', 'blue', 'yellow']

# First argument to pie() should be values array
# We can pass additional parameters for customizing the graph
# Different parameters are 
#   labels -> An array of labels for each division in the pie chart
#   colors -> An array of colors for each division in the pie chart
#   autopct -> An expression to show percentage value in each division of the pie chart. 
#              Let say expression is 
#                a) '%1.2f%%', this means percentage should be a floating point value with 2 decimals and %% for showing percentage symbol
#                b) '%1i%%', this means pecentage should be in integer format with percentage symbol
#                c) '%1f%%', this means percentage should be a floating point number with percentage symbol.
#   explode -> An array showing which division to explode in pie chart. Value of  less than 1 will explode, 0 will not explode
#   shadow -> A boolean value of True will show shadow
#   startangle -> A value <= 360 to rotate the pie chart
plt.pie(values, 
        labels=categories, 
        autopct="%1.2f%%", 
        colors=colors, 
        explode=[0.2, 0, 0, 0.1],
        shadow=True,
        startangle=180
       )
plt.show()
```
