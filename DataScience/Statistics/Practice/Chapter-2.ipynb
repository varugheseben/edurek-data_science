```python
# Variance and Standard Deviation
# Variance: In statistics and data science, variance is a value that describes how far a set of numbers is spread out from their average(mean) value.
#           It esssentially measures the 'volatility' or 'dispersion' with in a data set
#           Variance is calculated by taking the average of the squared differences from the Mean. 
#           Why do we square the differences?
#              If you simply added the differences from the mean, the positive and negative values would cancel each other out, resulting in zero. 
#              Squaring ensures all values are positive and gives more "weight" to outliers
#           Usage of Variance
#              Variance is a foundational concept used across various fields to understand risk, consistency, and data integrity
#              Used in 
#                  1) Risk Assessment in Finance
#                  2) Quality Control in Manufacturing
#                  3) Machine Learning and Data Science
#                      In Data Science, 
#                         High variance means model is "overfitting" (it follows the training data too closely and fails to generalize to new data).
#                         Low variance means model is stable and performs consistently across different datasets.
#                  4) Comparing Datasets
# Standard Deviation: It is the square root of the variance. It helps to check whether our data set has outlier or not.
#                     Low standard deviation means dataset is consistant and not far spread out from the mean.
#                     High standard deviation means dataset is not consitstent and spread far from the mean.
# While variance is the mathematical powerhouse behind the scenes, standard deviation is what we actually use for communication and visualization.
# Variance can be large value and its difficult to use for communication and visualization. Standard deviation will be smaller values and easy for 
# communication and visualization.

# Lets take an example of marks of two classes and determine which class is performing well.
import numpy as np
from scipy import stats

class_a = np.array([70, 72, 68, 71, 69])  # low spread
class_b = np.array([50, 90, 40, 95, 75])  # High spread

print(f'Mean class a: {np.mean(class_a)}')
print(f'Mean class b: {np.mean(class_b)}')

print(f"Standard Deviation class a {round(np.std(class_a), 2)}")
print(f"Standard Deviation class b {round(np.std(class_b), 2)}")

# So we have mean same for both classes so we cannot use mean value to determine the behavior of the dataset. That means by using mean value
# we cannot predict which class has consistently performing students. So for that purpose we can use standard deviation
# class_a has low standard deviation which means that students of class a is consistent
# class_b has high deviation which means that students of class b is not consistent. That means that dataset has some outlier
```
