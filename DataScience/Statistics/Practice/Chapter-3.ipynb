```python
# Box plot
# To draw box plot, mechanism is to divide given data set sorted first and divide them into 4 parts
# Lower Quartile (Q1) : Its first 25% of data sete and to find Q1 position we can use formula (n+1)/4
# To find the median for given set of size n we can use formule (n+1)/2. Ist 50% of data set
# Upper Quartile (Q3) : Its 75% of data et and to find Q3 position we can use formula as 3(n+1/4)
# Inter Quartile Range(IQR) = Q3 - Q1
# MInimum = Q1 - 1.5 * IQR
# Maximum = Q3 + 1.5 * IQR

import numpy as np
import matplotlib.pyplot as plt

data_set = np.array([23, 23, 24, 28, 29, 29, 30, 31, 33, 35, 36, 36, 37 ,38, 40, 40, 41, 42, 44, 45, 49, 53, 59])
Q1 = np.percentile(data_set,25)
median = np.percentile(data_set,50)
Q3 = np.percentile(data_set,75)
IQR = Q3 - Q1
print(f"Q1: {Q1}")
print(f"Median: {median}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Minimum: {Q1 - (1.5 * IQR)}")
print(f"maximum: {Q3 + (1.5 * IQR)}")
print("==============================================")

plt.boxplot(data_set)
plt.show()
```

```python
# We can use box plot for identifying the outliers
import numpy as np
import matplotlib.pyplot as plt

data_set = np.array([-10, -5, 0, 1, 5, 23, 23, 24, 28, 29, 29, 30, 31, 33, 35, 36, 36, 37 ,38, 40, 40, 41, 42, 44, 45, 49, 53, 59, 70, 75, 80])
Q1 = np.percentile(data_set,25)
median = np.percentile(data_set,50)
Q3 = np.percentile(data_set,75)
IQR = Q3 - Q1
minimum = Q1 - (1.5 * IQR)
maximum = Q3 + (1.5 * IQR)
print(f"Q1: {Q1}")
print(f"Median: {median}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Minimum: {minimum}")
print(f"maximum: {maximum}")
print("==============================================")

plt.boxplot(data_set)
plt.show()

# In the below graph, outliers are 
#            values < 0.5
#            values > 68.5

# Now we will remove the outlier from dataset
clean_data_set = data_set[(data_set >= minimum) & (data_set <= maximum) ]
Q1 = np.percentile(clean_data_set,25)
median = np.percentile(clean_data_set,50)
Q3 = np.percentile(clean_data_set,75)
IQR = Q3 - Q1
minimum = Q1 - (1.5 * IQR)
maximum = Q3 + (1.5 * IQR)
print(f"Q1: {Q1}")
print(f"Median: {median}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Minimum: {minimum}")
print(f"maximum: {maximum}")
print(f"Cleaned Data Set Trail 1: {clean_data_set}")
print("================================================================================================\n")
plt.boxplot(clean_data_set)
plt.show()

# We still see outlier, so we will redo the cleaning
clean_data_set = clean_data_set[(clean_data_set >= minimum) & (clean_data_set <= maximum) ]
Q1 = np.percentile(clean_data_set,25)
median = np.percentile(clean_data_set,50)
Q3 = np.percentile(clean_data_set,75)
IQR = Q3 - Q1
minimum = Q1 - (1.5 * IQR)
maximum = Q3 + (1.5 * IQR)
print(f"Q1: {Q1}")
print(f"Median: {median}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Minimum: {minimum}")
print(f"maximum: {maximum}")
print(f"Cleaned Data Set Trail 2: {clean_data_set}")
print("================================================================================================\n")
plt.boxplot(clean_data_set)
plt.show()
```
