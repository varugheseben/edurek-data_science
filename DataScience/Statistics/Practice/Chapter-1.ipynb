```python
# Measure of central tendency
# Three different values are important for this analysis, Mean, Median and Mode
# Mean is the average of data set
# Median: First we need to sort the list in ascending order
#         Determine the count of numbers, 
#               if its even then median will be average of 2 middle numbers
#               if its odd then median will be middle value in the list
# Mode: It will be the value which is repeated most. If there is none repeating the it will use the first value in the list
import numpy as np
from scipy import stats

salaries = np.array([45000, 50000, 52000, 53000, 55000, 200000])

print(f'Mean: {np.mean(salaries)}')
print(f'Median: {np.median(salaries)}')
print(f'Mode: {stats.mode(salaries).mode}')
```
