```python
# Example of filtering using panda
import pandas as pd

df = pd.read_csv("Practice/data/pokemon.csv")
print(df)
```

```python
# Filter pokemons which has height >= 2
tall_pokemon = df[df["Height"] >= 2]
print(tall_pokemon)
```

```python
# Filter pokemons which has weight> 100
heavy_pokemon = df[df["Weight"] > 100]
print(heavy_pokemon)
```

```python
# Filter pokmons which are legendary
legendary_pokemons = df[df["Legendary"] == 1]
print(legendary_pokemons)
```

```python
# Example of using logical operator.
# Note: When we use multiple condition, put each condition in a ()
water_pokemon = df[
                    (df['Type1'] == 'Water') |
                    (df['Type2'] == 'Water')
                ]
print(water_pokemon)
```

```python
ff_pokemon = df[
    (df['Type1'] == 'Fire') &
    (df['Type2'] == "Flying")
]
print(ff_pokemon)
```
