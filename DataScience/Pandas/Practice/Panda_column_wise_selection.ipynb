```python
# Selecting data from dataframe. There are 2 ways for it
# 1) Column wise selection
# 2) Row wise selection

# Column wise selection
import pandas as pd

df = pd.read_csv("data/world_literacy.csv")
# Bydefaul panda will print only first 5 and last 5 rows
print(df['Entity'])
```

```python
# To print the entire column set use to_string()
import pandas as pd

df = pd.read_csv("data/world_literacy.csv")
print(df['Entity'].to_string())
```

```python
# Selecting multiple columns
import pandas as pd

df = pd.read_csv("data/world_literacy.csv")
print(df[["Entity", "Code", "Year"]])
```

```python
# Selecting multiple columns and entire set
import pandas as pd

df = pd.read_csv("data/world_literacy.csv")
print(df[["Entity", "Code", 'Year']].to_string())
```
