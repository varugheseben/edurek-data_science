```python
# Panda has library function for cleaning the data. 
# Data cleaning is the process of removing/ fixing the incomplete, inconsistent and irrelevant data
```

```python
# Dropping a column
import pandas as pd

pokemons = pd.read_csv('Practice/data/pokemon.csv')

# Remove irrelevant columns
pokemons = pokemons.drop(columns=['Legendary', 'No']) # This will remove two columns
print(pokemons)
```

```python
# Handle removing of missing data
pokemons = pokemons.dropna() # This will remove rows if any of the column value is missing
print(pokemons)
```

```python
# Handle removing of missing data for selected columns
pokemons = pd.read_csv('Practice/data/pokemon.csv')
pokemons = pokemons.dropna(subset=["Type2"]) # Here removing based of column Type2
print(pokemons)
```

```python
# Fill missing data with meaning full data
import pandas as pd

pokemons = pd.read_csv('Practice/data/pokemon.csv')
pokemons = pokemons.fillna({"Type1": -1, "Type2":-2}) # Here if value for columns Type1 and Type2 are missing then its replaced with -1 &-2 respectively
print(pokemons)
```

```python
# Fix inconsistent values
import pandas as pd

pokemons = pd.read_csv("Practice/data/pokemon.csv")
# lets say we wanted to replace value of Type1 column with some consistent values
pokemons['Type1'] = pokemons['Type1'].replace({'Grass': 'GRASS'}) # This will replace all occurance of 'Grass' with 'GRASS' for column Type1
# We can pass multiple key value pair to replace()
print(pokemons)
```

```python
# Standardizing text column values
pokemons['Name'] = pokemons['Name'].str.lower()
print(pokemons)
```

```python
# Fix data types
import pandas as pd

pokemons = pd.read_csv('Practice/data/pokemon.csv')
pokemons['Legendary'] = pokemons['Legendary'].astype(bool) # This basically converted the data type of column 'Legendary' to boolean
print(pokemons)
```

```python
# Removing duplicates. 
import pandas as pd

pokemons = pd.read_csv("Practice/data/duplicate_pokemon.csv")
print(pokemons.to_string())
print("---------------------------------------------------------------")
pokemons = pokemons.drop_duplicates()
# If we not passed a subset to drop_duplicates() then duplicate removal will be based on all columns
print(pokemons.to_string())
```

```python
# Dropping duplicates based on a subset
import pandas as pd

pokemons = pd.read_csv('Practice/data/duplicate_pokemon.csv')
print(pokemons.to_string())
print("---------------------------------------------------------------")
pokemons = pokemons.drop_duplicates(subset=['Type1','Type2'])
# Here we passed subset and duplicate removal will be based on that subset
print(pokemons.to_string())
```
