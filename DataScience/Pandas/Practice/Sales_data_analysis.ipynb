```python
# If the Price column has missing values, how do you fill them with the median price of the entire column?
sales_data = pd.read_csv('data/sales_data.csv')
print(sales_data)
price_median = sales_data['Price'].median()
print(f'Price Median: {price_median}')
sales_data['Price'] = sales_data['Price'].fillna(price_median)
sales_data
# Why Use Median Instead of Mean?
# While dealing with prices median is used over mean because
# 1) It is robust to outliers. 
#   That means imagine 5 people in a room earning 50000/year. So mean = 50000 and median = 50000.
#   Lets say billionare enters the room, then the income of that person will be an outlier compared to others. 
#   So value mean will become billions of dollars but median remains 50000. 
#   That means mean is sensitive to outliers but median is robust to outliers.
# 2) It ensures the filled values are more "typical" of your existing data. 
#    In data science, "typical" means that the new values shouldn't be "weird" or "extreme" compared to what is already there. 
#    They should blend into the existing distribution so they don't skew your final analysis.
```

```python
# Fill multiple numeric columns with their respective medians for missing values
sales_data = pd.read_csv('data/sales_data.csv')
print(sales_data)
sales_data = sales_data.fillna(sales_data.median(numeric_only=True))
sales_data
```

```python
# Drop all rows where the Product name is null?
sales_data = pd.read_csv('data/sales_data.csv')
print(sales_data)
sales_data = sales_data.dropna(subset=['Product'])
print(sales_data)
```

```python
# Convert the Date column from a string (object) type to a proper datetime object.
initial_type = type(sales_data['Date'].iloc[0])
print(f'Before conversion : {initial_type}')
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
converted_type = type(sales_data['Date'].iloc[0])
print(f'After conversion : {converted_type}')
```

```python
# Find the total revenue (Price * Quantity) for each Category.
# Create a new column 'Revenue' in sales_data
sales_data['Revenue'] = sales_data['Price'] * sales_data['Quantity']

revenue_by_category = sales_data.groupby('Category')
print(revenue_by_category['Revenue'].sum())
```

```python
# Count how many unique products are sold in each category.
group_by_category = sales_data.groupby('Category')
print("Count using nunique()")
print(group_by_category['Product'].nunique())
print("=================================================")
print("Count using count()")
print(group_by_category['Product'].count()) # This will not be correct
#count(): Counts every row in that category (including duplicates). If "Apple" appears 5 times, it counts as 5.
#nunique(): Counts only the distinct names. If "Apple" appears 5 times, it counts as 1.
```

```python
# Get the maximum and minimum price for each category in a single command using .agg().
group_by_category = sales_data.groupby('Category')
price_by_category = group_by_category['Price']
print(price_by_category.agg(['min', 'max']))

# We can do this in single command as
sales_data.groupby('Category')['Price'].agg(['min', 'max'])
```

```python
# Create a new column called Discount_Price that is 90% of the original Price.
sales_data = pd.read_csv('data/sales_data.csv')
sales_data['Discount_Price'] = sales_data['Price'] * (90/100)
sales_data
```

```python
# Creating column using assign(). Assign will add new column and return a new data frame so your original data frame is not
# changed
sales_data = pd.read_csv('data/sales_data.csv')
new_sales_data = sales_data.assign(Discount_price = sales_data['Price'] * (90/100))
print(new_sales_data) # New column is added
print('---------------------------------------------------------')
print(sales_data)
```

```python
# Apply discount to a specific category
import numpy as np
sales_data = pd.read_csv('data/sales_data.csv')
new_sales_data = sales_data
new_sales_data['Discount_Price'] = np.where(sales_data['Category'] == 'Electronics', 
                                            sales_data['Price'] * (90/100),
                                            sales_data['Price']
                                           )

# Discount price is calculated only for 'Electronics' category for other categories it will be same as price value
print(new_sales_data[sales_data['Category'] == 'Electronics']) 
print('----------------------------------------------')
print(new_sales_data)
```
