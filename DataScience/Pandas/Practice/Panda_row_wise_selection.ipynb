```python
# Row wise selection using panda
import pandas as pd

df = pd.read_csv("data/world_literacy.csv")
# We can use loc[] for fetching individual rows
print(df.loc[0])
```

```python
# We can label index by values of a column
import pandas as pd

df = pd.read_csv("data/world_literacy.csv", index_col="Entity")
print(df)
print(df.loc["United States"])
```

```python
# We can do selection on columns we want on each row
import pandas as pd

df = pd.read_csv("data/world_literacy.csv", index_col="Entity")
print(df.loc["United States", ["Year", "Literate"]])
```

```python
# We can do range on labels
import pandas as pd

df = pd.read_csv("data/world_literacy.csv", index_col="Entity")
print(df.loc["United States":"Vantau", ["Year", "Literate"]])
```
