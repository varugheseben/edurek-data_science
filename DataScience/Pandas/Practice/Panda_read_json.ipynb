```python
# Example of reading from JSON file
import pandas as pd

# By default, Pandas will read first 5 rows & columns and last 5 rows & columns
df = pd.read_json("data/example.json")
print(df)
```

```python
# Printing the entire data of a json file
import pandas as pd

df = pd.read_json("data/example.json")
print(df.to_string())
```
