```python
# Panda has aggregate functions which will allow us to find mean, sum, min, max and count of dataframe
import pandas as pd

pokemons = pd.read_csv("data/pokemon.csv")
# mean() to find the mean of numeric column values
print(pokemons.mean(numeric_only=True))
```

```python
# sum() to find the sum of numeric column values
print(pokemons.sum(numeric_only=True))
```

```python
# min() and max() to find minimum and maximum from column values
print(pokemons.min(numeric_only=True))
print('--------------------------------------------------------')
print(pokemons.max(numeric_only=True))
```

```python
# count() for getting the count of each column values
print(pokemons.count())
```

```python
# Above functions are used with entire data frame.
# We can also use aggregate function with columns

print(f"Mean of Height: {pokemons['Height'].mean()}")  # Here we dont need to pass the keyword argument as 'Height' column is numeric
print(f"Sum of Height: {pokemons['Height'].sum()}")  # Here we dont need to pass the keyword argument as 'Height' column is numeric
print(f"Min of Height: {pokemons['Height'].min()}")  # Here we dont need to pass the keyword argument as 'Height' column is numeric
print(f"Max of Height: {pokemons['Height'].max()}")  # Here we dont need to pass the keyword argument as 'Height' column is numeric
print(f"Count of Height: {pokemons['Height'].count()}")  # Here we dont need to pass the keyword argument as 'Height' column is numeric
```

```python
# Another function is groupby. Like SQL we can group by data based on column
import pandas as pd

pokemons = pd.read_csv("data/pokemon.csv")
group = pokemons.groupby('Type1')
print(group)

# Now we can use this groupby object to perform different aggregare functions
print(group["Height"].mean())
```

```python
print(group['Height'].sum())
```

```python
print(group['Height'].min())
```

```python
print(group['Height'].max())
```

```python
print(group['Height'].count())
```
