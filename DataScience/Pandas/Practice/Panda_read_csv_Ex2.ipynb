```python
# value_counts() : This function is used for getting unique count by value of a given column
# This is method of Series not the DataFrame
import pandas as pd

data_frame = pd.read_csv('data/matches.csv')
data_frame['winner'].value_counts()
```

```python
# Get the winner counts for data where city = 'Mumbai'
data_frame[data_frame['city'] == 'Hyderabad']['winner'].value_counts()
```

```python
data_frame.value_counts()
```

```python
# Write expression to list out total games played by each team
data_frame['team1'].value_counts() + data_frame['team2'].value_counts()
```

```python
# sort_values() : This function is used for sorting values in ascending order
# So write expression to list out total games played by each team in ascending order
(data_frame['team1'].value_counts() + data_frame['team2'].value_counts()).sort_values()
```

```python
# To sort in descending order use function as sort_value(ascending=False)
# So write expression to list out total games played by each team in descending order
(data_frame['team1'].value_counts() + data_frame['team2'].value_counts()).sort_values(ascending=False)
```
