```python
# Series is 1D labelled data, either single row or single column
# Example of series in panda
# Integer data
import pandas as pd

data_1 = [1,2,3,4,5,6]
series_1 = pd.Series(data_1)
print(series_1)
```

```python
# Floating point data
import pandas as pd

data_1 = [1.5, 2.5, 3.5, 4.5]
series_1 = pd.Series(data_1)
print(series_1)
```

```python
# String data
import pandas as pd

data_1 = ['A', 'B', 'C', 'D']
series_1 = pd.Series(data_1)
print(series_1)
```

```python
# Boolean data
import pandas as pd

data_1 = [True, True, False, True]
series_1 = pd.Series(data_1)
print(series_1)
```

```python
# In all above examples, we have 'index' and 'value' printed from the series. And its shows dtype as data type.
# Bydefault, index is always numeric and will start from 0
# We can also have our own index value being used, for that purpose we need to pass a keyword argrument to Series()
import pandas as pd

data_1 = [1,2,3,4,5,6]
series_1 = pd.Series(data_1, index=[1,2,3,4,5,6])
print(series_1)
```

```python
# Using labelling for Series index
import pandas as pd

data_1 = [1,2,3,4,5,6]
series_1 = pd.Series(data_1, index=['A', 'B', 'C', 'D', 'E', 'F'])
print(series_1)
```

```python
# Using dictionaries to create Panda Series
# In the case of dictionary data keys will be used for labelling index
import pandas as pd

calories = {
    'Day-1': 1750,
    'Day-2': 2100,
    'Day-3': 1800,
    'Day-4': 2300,
    'Day-5': 1850
}
series_1 = pd.Series(calories)
print(series_1)
```

```python
# Accessing the individual elements from a Series
# We have two functions we can use for accessing elements
#  1) loc  -> This case we need to pass the label as the subscript
#  2) iloc -> This case we need to pass numerical index as the subscript
import pandas as pd

calories = {
    'Day-1': 1750,
    'Day-2': 2100,
    'Day-3': 1800,
    'Day-4': 2300,
    'Day-5': 1850
}
series_1 = pd.Series(calories)
print(f"Using loc: {series_1.loc['Day-2']}")
print(f"Using loc: {series_1.loc['Day-4']}")
print(f'Using iloc: {series_1.iloc[0]}')
print(f'Usinf iloc: {series_1.iloc[2]}')
```

```python
# Filtering from Panda series
import pandas as pd

calories = {
    'Day-1': 1750,
    'Day-2': 2100,
    'Day-3': 1800,
    'Day-4': 2300,
    'Day-5': 1850
}
series_1 = pd.Series(calories)
# Filter calories greater than 2000
unhealthy_diet = series_1[series_1 > 2000]
print(unhealthy_diet)
print('-----------------------------------------')

# Filter calories lesser than or equal to 2000
healthy_diet = series_1[series_1 <= 2000]
print(healthy_diet)
```
