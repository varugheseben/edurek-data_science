```python
# Dataframe is a 2D labelled data structure which will have data arranged in rows and columns
# DataFrame in Pandas help to work with tabular data, data which is in rows and column format
import pandas as pd

employees = {
    "Name": ["Denny", "Mariza", "Joshua", "Jonathan", "Jacob", "Benny"],
    "Age": [45, 18, 17, 16, 17, 43],
}

data_frame = pd.DataFrame(employees)
print(data_frame)
```

```python
# We can add label to the index of dataframe
import pandas as pd

employees = {
    "Name": ["Denny", "Mariza", "Joshua", "Jonathan", "Jacob", "Benny"],
    "Age": [45, 18, 17, 16, 17, 43],
}
data_frame = pd.DataFrame(employees, index=["Employee 1", "Employee 2", "Employee 3", "Employee 4", "Employee 5", "Employee 6"])
print(data_frame)
```

```python
# Fetch individual value from a dataframe. To do so we can use loc and iloc
import pandas as pd

employees = {
    "Name": ["Denny", "Mariza", "Joshua", "Jonathan", "Jacob", "Benny"],
    "Age": [45, 18, 17, 16, 17, 43],
}
data_frame = pd.DataFrame(employees, index=["Employee 1", "Employee 2", "Employee 3", "Employee 4", "Employee 5", "Employee 6"])
print(f"Using loc \nEmployee 1: {data_frame.loc['Employee 1']}")
print("============================================================")
print(f"Using iloc \nEmployee 1: {data_frame.iloc[0]}")
```

```python
# Adding a new column to dataframe
import pandas as pd

employees = {
    "Name": ["Denny", "Mariza", "Joshua", "Jonathan", "Jacob", "Benny"],
    "Age": [45, 18, 17, 16, 17, 43],
}
data_frame = pd.DataFrame(employees, index=["Employee 1", "Employee 2", "Employee 3", "Employee 4", "Employee 5", "Employee 6"])
data_frame['Hobbies'] = ['Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family']
print(data_frame)
```

```python
# Adding new row to dataframe
import pandas as pd

employees = {
    "Name": ["Denny", "Mariza", "Joshua", "Jonathan", "Jacob", "Benny"],
    "Age": [45, 18, 17, 16, 17, 43],
}
data_frame = pd.DataFrame(employees, index=["Employee 1", "Employee 2", "Employee 3", "Employee 4", "Employee 5", "Employee 6"])
data_frame['Hobbies'] = ['Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family']

# We create a new data frame and then concatenate with original one
employee = {
    "Name": "Varughese",
    "Age": 75,
    "Hobbies": "Spend time with Family"
}
new_row = pd.DataFrame([employee], index=['Employee 7'])
data_frame = pd.concat([data_frame, new_row])
print(data_frame)
```

```python
# Adding multiple rows to dataframe
import pandas as pd

employees = {
    "Name": ["Denny", "Mariza", "Joshua", "Jonathan", "Jacob", "Benny"],
    "Age": [45, 18, 17, 16, 17, 43],
}
data_frame = pd.DataFrame(employees, index=["Employee 1", "Employee 2", "Employee 3", "Employee 4", "Employee 5", "Employee 6"])
data_frame['Hobbies'] = ['Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family']

# When we need to add multiple rows the pass array of dictionary to DataFrame and then concatenate that dataframe with original one
new_employees = [
    {
        "Name": "Varughese",
        "Age": 75,
        "Hobbies": "Spend time with Family"
    },
    {
        "Name": "Lissy",
        "Age": 65,
        "Hobbies": "Spend time with Family",
    }
]

new_frames = pd.DataFrame(new_employees, index=["Employee 7", "Employee 8"])
data_frame = pd.concat([data_frame, new_frames])
print(data_frame)
```

```python
# head() and tail()
import pandas as pd

employees = {
    "Name": ["Denny", "Mariza", "Joshua", "Jonathan", "Jacob", "Benny"],
    "Age": [45, 18, 17, 16, 17, 43],
}
data_frame = pd.DataFrame(employees, index=["Employee 1", "Employee 2", "Employee 3", "Employee 4", "Employee 5", "Employee 6"])
data_frame['Hobbies'] = ['Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family', 'Spend time with Family']

# When we need to add multiple rows the pass array of dictionary to DataFrame and then concatenate that dataframe with original one
new_employees = [
    {
        "Name": "Varughese",
        "Age": 75,
        "Hobbies": "Spend time with Family"
    },
    {
        "Name": "Lissy",
        "Age": 65,
        "Hobbies": "Spend time with Family",
    }
]

new_frames = pd.DataFrame(new_employees, index=["Employee 7", "Employee 8"])
data_frame = pd.concat([data_frame, new_frames])
data_frame.head() # Returns first 5 rows
```
