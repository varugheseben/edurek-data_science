```python
# Applying Bayes theorem
# Lets say we have 10000 emails
#   In that 10% detected as spam
#   We have 90% accuracy in spam detection
#   We have 10% mails which are wrongly marked as spam
#   So we can say components of Bayes theorem as
#      So we have 2 events
#        1) Spam emails received as event A
#        2) Detecting spam emails as event B
#      So we can write
#        P(A) = 0.10
#        P(Not A) = 0.90
#        P(B|A) = 0.90
#        P(B|Not A) = .10
#      So P(A|B) = (0.90 * 0.10) / ((0.90 * 10) + (0.10 * 0.90)
#                = 0.09/(0.09 + 0.09)
#                = 0.09/0.18 = 0.5
# We can explain the scenario as 
#    We have 10000 email received. In that 10% are spam that means we have 1000 spam emails.
#    And we have assumption model that 90% of spam detection is correct and 10% are falsely detected
#    So from 1000 spam emails we can say 900 are truely spam and 100 are missed
#    Also we know that there are 9000 email are not spam but in that 10 are falsely spammed so that will be 900
#    So total spams detected by assumption model is 900 + 900 = 1800
#    So probability of an email being detected as spam is 900/1800 = 0.5 = 50%

# Below is a example of application of Bayes theorem to above scenario
```

```python
import pandas as pd

total_emails = 10000
spam_emails_ratio = 0.10 # P(A)
not_spam_emails_ratio = 0.90 # P(Not A)
true_positive_ratio = 0.90 # P(B|A)
false_positive_ratio = 0.10 # P(B| Not A)

# ACTUAL COUNTS
spam = int(total_emails * spam_emails_ratio) # This will be 1000
not_spam = int(total_emails - spam) # This will be 9000

# Assumption model
# Actual spam emails correctly flagged as spam.
true_positive = int(spam * true_positive_ratio) # This will be 1000 * 0.90 = 900. This will P(B|A)*P(A)

# Legitimate emails mistakenly flagged as spam.
false_positive = int(not_spam * false_positive_ratio) # This will be 9000 * 0.10 = 900. This will be P(B|Not A) * P(Not A)

# Actual spam emails that slipped through to the inbox.
false_negative = int(spam - true_positive) # This will be 1000 - 900 = 100

# Legitimate emails correctly left alone.
true_negative = int(not_spam - false_positive) # This will be 9000 - 900 = 8100

data = {
    "Actual Spam" : [true_positive, false_negative],
    "Actual Not Spam" : [false_positive, true_negative]
}

data_frame = pd.DataFrame(data, index=['Predicted Spam', 'Predicted non Spam'])

# Applying Bayes theorem for calculating probabilty
probability = true_positive / (true_positive + false_positive)
print(f'Probability: {probability}')
```
