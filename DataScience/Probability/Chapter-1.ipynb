```python
# Simulation of rolling a dice
# Find the probability of getting even number when dice is rolled

import numpy as np

data_set = np.random.randint(1, 7, 6)
print(data_set)
favorable_outcomes = np.sum(data_set % 2 == 0)
total_outcomes = len(data_set)

probability = (favorable_outcomes / total_outcomes) * 100

print(f"Probability of getting even numbers: {probability}")
```
