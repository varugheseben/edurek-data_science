```python
# Generate data set for normal distribution
# Normal distribution will be a data set where values are spread near to the average
import numpy as np

data_set = np.random.normal(loc=50, scale=1, size=10)
# loc is for mean
# scale is for standard deviation
# size is for total number of data set need to be generated
print(data_set)
# In below data set we have values which are very close to 50
```

```python
# Generate data set for normal distribution
# Normal distribution will be a data set where values are spread near to the average
import numpy as np

data_set = np.random.normal(loc=50, scale=10, size=10)
# loc is for mean
# scale is for standard deviation
# size is for total number of data set need to be generated
print(data_set)
# In below example we have values which are little far apart from the mean. This is because we passed large value for standard deviation
```

```python
# Generate a normal distribution dataset and analyze the probability of event 
# For that purpose we will generate a 10000 data set
# For analyzing the probability we will plot histogram

import numpy as np
import matplotlib.pyplot as plt

data_set = np.random.normal(loc=50, scale=10, size=10000)
plt.hist(data_set)
plt.title("Normal Distribution")
plt.show()

# Lets say below graph indicates spending trend of people of different ages.
# So X-axis indicates ages and Y axis indicates amount spending.
# In the below graph we can see x-axis value of range 45 -50 has high values. That means that data range has maximum probability of occurance.
# Basically this approach is usefull in determining the probabilty of an event
# So with this graph we can say people of ages between 45 -50 has high probability of spending large amounts
```
