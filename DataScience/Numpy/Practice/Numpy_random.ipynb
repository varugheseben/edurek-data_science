```python
# Generating random integers
import numpy as np

range_generator = np.random.default_rng()
print(range_generator.integers(1,7))
print('--------------------------------')
print(range_generator.integers(low=2, high=9))
print('--------------------------------')
print(range_generator.integers(3,8, size=3))
print('--------------------------------')
print(range_generator.integers(3,8, size=(3,2)))
```

```python
# Setting seed for generating random numbers
# When seed is set then it will generate the same numbers
import numpy as np

range_generator = np.random.default_rng(seed=5)
array = np.array([10, 5, 4, 8, 11, 23, 27, 30, 45])
print(range_generator.integers(1, 10, size=3))
print(range_generator.integers(1, 10, size=3))
print(range_generator.integers(1, 10, size=3))
print("-------------------------------------------")
# Reason why different print gives different result is because we are using same range_generator for all 3 prints.
# So once a seed is set, then it will give different numbers every time we generate numbers. If we want same result every time,
# then we need to intiate the generator every time with same seed like below
range_generator = np.random.default_rng(seed=5)
print(range_generator.integers(1, 10, size=3))
range_generator = np.random.default_rng(seed=5)
print(range_generator.integers(1, 10, size=3))
range_generator = np.random.default_rng(seed=5)
print(range_generator.integers(1, 10, size=3))
```

```python
# Generating random floating point numbers
import numpy as np

random_number = np.random.uniform()
print(random_number)
print('------------------------------')
random_number = np.random.uniform(-1, 1)
print(random_number)
print('------------------------------')
random_number = np.random.uniform(low=-1, high=1)
print(random_number)
print('------------------------------')
random_numbers = np.random.uniform(-1, 1, size=3)
print(random_numbers)
print('------------------------------')
random_array = np.random.uniform(-1, 1, size=(3,2))
print(random_array)                
```

```python
# Setting seed for generating random numbers
# Similarly as in integers if we want same number to be generated then we have to reinitiate the seed() 
import numpy as np

np.random.seed(seed=5)
random_number = np.random.uniform()
print(random_number)
print('------------------------------')
random_number = np.random.uniform()
print(random_number)
print('------------------------------')
random_number = np.random.uniform(-1, 1)
print(random_number)
print('------------------------------')
random_number = np.random.uniform(low=-1, high=1)
print(random_number)
print('------------------------------')
random_numbers = np.random.uniform(-1, 1, size=3)
print(random_numbers)
print('------------------------------')
random_array = np.random.uniform(-1, 1, size=(3,2))
print(random_array)                
```

```python
# Shuffling array using random generator
import numpy as np

random_generator = np.random.default_rng()
array = np.array([1, 3, 5, 7, 9, 11, 13])
print(f'Original Array: {array}')
random_generator.shuffle(array)
print(f'After first shuffle: {array}')
random_generator.shuffle(array)
print(f'After second shuffle: {array}')
```

```python
# Choosing random elements from a array
import numpy as np

range_generator = np.random.default_rng()
fruits = np.array(['apple', 'banana', 'cantalope', 'papaya', 'blueberry'])
choice = range_generator.choice(fruits)
print(f'Fruit chosen: {choice}')
choices = range_generator.choice(fruits, size=(4,4))
print(f'Fruits chosen: {choices}')
```
