```python
# Given a 10X10 matrix of ones, change all the interior values (not on the border) to 0.
import numpy as np

array = np.ones((10,10), dtype=int)
print(array)
print('--------------------------------')
array[1:-1,1:-1] = 0 # 1:Starts the selection at the second row/column (skipping the first border).
                     # -1: Ends the selection just before the last row/column (skipping the last border).
print(array)
```

```python
# Create a 5X5 matrix with values ranging from 0 to 24 and extract the second and fourth rows.
import numpy as np

array = np.arange(0,25)
print(array)
print('------------------------------------------------')
array_5X5 = array.reshape(5,5)
print(array_5X5)
print('------------------------------------------------')
print(f'Second Row: {array_5X5[1]}')
print(f'Fourth Row: {array_5X5[3]}')
```

```python
# Create a random 1D array of size 20. 
# Replace all values greater than 0.5 with 1 and all values less than or equal to 0.5 with 0.
import numpy as np

random_array = np.random.rand(20)
print(random_array)
print('-------------------------------------------------')

# Replace values > 0.5 with 1 and values <= 0.5 with 0
processed_array = np.where(random_array > 0.5, 1, 0)
print("\nProcessed Array:")
print(processed_array)
```
