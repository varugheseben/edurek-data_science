```python
# Create a 3X3 identity matrix.
# In linear algebra, an identity matrix is a square matrix with ones on the main diagonal and zeros everywhere else.
import numpy as np

identity_3X3 = np.eye(3)
print(identity_3X3)
```

```python
# Create a 1D array of numbers from 10 to 49
import numpy as np

range_generator = np.random.default_rng(seed=10)
array  = np.array(range_generator.integers(10,50, size=(40,))) # this method can contain repeated elements
print(array)
print('------------------------------------------------')
# We also can use arange()
array = np.arange(10, 50)
print(array)
```

```python
# Reverse a vector or numpy array
import numpy as np

array = np.arange(10, 50)
reverse_1 = array[::-1]
print(reverse_1) # Fastest
print('--------------------------------------------------')
reverse_2 = np.flip(array)
print(reverse_2)
print('--------------------------------------------------')
reverse_3 = np.flipud(array)
print(reverse_3)
```

```python
# Create an 8 X 8 matrix and fill it with a checkerboard pattern (alternating 0s and 1s).
import numpy as np

# 1. Initialize an 8x8 matrix of zeros
checkerboard = np.zeros((8, 8), dtype=int)
print(checkerboard)
print('--------------------------------------')

# 2. Fill specific rows and columns with 1s
# Pattern: [start_index:end_index:step]

# Target rows 1, 3, 5, 7 and columns 0, 2, 4, 6
checkerboard[1::2, ::2] = 1
print(checkerboard)
print('--------------------------------------')

# Target rows 0, 2, 4, 6 and columns 1, 3, 5, 7
checkerboard[::2, 1::2] = 1

print(checkerboard)
```

```python
# Create an 8 X 8 matrix and fill it with a checkerboard pattern (alternating 0s and 1s) using tile()
block = np.array([[0, 1], [1, 0]])
checkerboard = np.tile(block, (4, 4))
print(checkerboard)
```

```python
# Create a 5X5 array with random values
import numpy as np

range_generator = np.random.default_rng(seed=10)
array  = np.array(range_generator.integers(1,50, size=(5,5))) # this method can contain repeated elements
print(array)
```

```python
# Create a 5X5 array with random values sampled from a standard normal distribution.
import numpy as np

rng = np.random.default_rng()
array = rng.standard_normal((5,5))
print(array)
```

```python
# Create 5X3 matrix
import numpy as np

print('Method 1')
array = np.arange(15).reshape(5,3)
print(array)
print('---------------------------------------')
print('Method 2')
array = np.random.randint(0, 10, (5, 3)) # Number range from 0 - 9
print(array)
```
