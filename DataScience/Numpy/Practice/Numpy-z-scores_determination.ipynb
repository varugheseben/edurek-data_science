```python
# Findout outliers using z-scores
# A z-score (also known as a standard score) measures exactly how many standard deviations a specific data point is away from 
# the mean (average) of a dataset.
# It is a powerful tool in data science and statistics for identifying outliers, normalizing data before feeding it into 
# machine learning models (like KNN or SVM), and comparing data points from different distributions.

import numpy as np

# 1. Create a sample dataset (50 normal values + 3 manual outliers)
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=50)
data = np.append(data, [75, 20, 80])  # Injecting extreme values/outliers

# 2. Calculate the mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# 3. Calculate z-scores using NumPy broadcasting
z_scores = (data - mean) / std_dev

# 4. Define a threshold for outliers (typically 2.5 or 3.0)
threshold = 2.5

# 5. Identify values and indices of the outliers
outlier_mask = np.abs(z_scores) > threshold
outlier_values = data[outlier_mask]
outlier_indices = np.where(outlier_mask)[0]

# Print Results
print(f"Dataset Mean: {mean:.2f}")
print(f"Dataset Standard Deviation: {std_dev:.2f}\n")

print("Identified Outliers:")
for idx, val in zip(outlier_indices, outlier_values):
    print(f"Index: {idx}, Value: {val:.2f}, Z-score: {z_scores[idx]:.2f}")
```
