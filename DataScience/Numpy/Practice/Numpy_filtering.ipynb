```python
# Numpy allows to filter the data set very efficiently
import numpy as np

ages = np.array([[25, 15, 18, 20, 31, 28, 45], [55, 16, 17, 24, 75, 60, 40]])
print(ages)
```

```python
# Filter all students with age < 18
teenages = ages[ages < 18]
print(teenages)
```

```python
# Filter all students with age > 18 but less than 50
adults = ages[(ages >= 18) & (ages < 50)]
print(adults)
```

```python
# Filter all students with age > 50
seniors = ages[ages >= 50]
print(seniors)
```

```python
# Above method is called as boolean indexing and it doesnt preserve the shape of the array.
# To preserve the shape of the array we can use 'where()'
import numpy as np

ages = np.array([[25, 15, 18, 20, 31, 28, 45], [55, 16, 17, 24, 75, 60, 40]])
teenages = np.where(ages < 18, ages, 0)
print(teenages)
print('-----------------------------------')

adults = np.where(((ages >= 18) & (ages < 50)), ages, 0)
print(adults)
print('------------------------------------')

seniors = np.where(ages > 50, ages, 0)
print(seniors)
```
