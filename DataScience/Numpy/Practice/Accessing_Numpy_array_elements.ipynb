```python
# Elements of a numpy array can be accessed in 2 ways
# 1) Using Chain indexing
# 2) Using Multi dimensional indexing
```

```python
# Chain Indexing
import numpy as np
array = np.array([[[1, 2, 3]], [[4, 5, 6]]])
print(array[0][0][0]) # will print as 1
print(array[0][0][2]) # will print as 3
print(array[1][0][2]) # will print as 6
# This is called chain indexing
print(array[2][0][2]) # will cause IndexError exception
```

```python
# Multi-dimensional indexing
import numpy as np
array = np.array([[[1, 2, 3]], [[4, 5, 6]]])
print(array[0,0,0]) # will print as 1
print(array[0,0,2]) # will print as 3
print(array[1,0,2]) # will print as 6
# This is called Multi-dimensional indexing
print(array[2,0,2]) # will cause IndexError exception
```
