```python
# Example of training labelled data(Supervised Learning)
# We will use scikit-learn to build a Classification model.
# In this example, we will create a mock labeled dataset of houses, where the features are "Square Footage" and "Number of Bedrooms",
# and the label is whether the house is "Expensive" (1) or "Affordable" (0).

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. Create a mock Labeled Dataset
# Features: [Square Footage, Bedrooms] -> Label: [1 for Expensive, 0 for Affordable]
data = {
    'Square_Footage': [1500, 2500, 1200, 3500, 1100, 4000, 1800, 3100],
    'Bedrooms': [3, 4, 2, 5, 2, 5, 3, 4],
    'Is_Expensive': [0, 1, 0, 1, 0, 1, 0, 1]  # <--- This is our target 'Label'
}

# Convert to a Pandas DataFrame for easy viewing
df = pd.DataFrame(data)
print("--- Our Labeled Dataset ---")
print(df)
print("\n" + "="*40 + "\n")

# 2. Separate Features (X) from the Label (y)
X = df[['Square_Footage', 'Bedrooms']]  # Features
y = df['Is_Expensive']                  # Target Label

# 3. Split the data into Training and Testing sets
# We use 75% of the data to train the model, and reserve 25% to test it.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Initialize the Supervised Learning Algorithm
# We are using a Random Forest Classifier
algorithm = RandomForestClassifier(max_depth=3, random_state=42)

# 5. TRAIN THE MODEL (This is where the algorithm learns from the labeled data)
# The .fit() function maps the features (X_train) to the correct answers (y_train)
trained_model = algorithm.fit(X_train, y_train)
print("Model training complete!")

# 6. Test the Model's Predictions
# We feed the model unseen features (X_test) and see what it guesses
predictions = trained_model.predict(X_test)

# Calculate how accurate the model was by comparing predictions to actual labels (y_test)
accuracy = accuracy_score(y_test, predictions)

print(f"Model Accuracy on Test Data: {accuracy * 100}%")
print(f"Actual Labels:     {list(y_test)}")
print(f"Model Predictions: {list(predictions)}")
```
