```python
# View the decision tree generated by DecisionTreeRegressor
# Training based on decision tree regression. In this case algorithm is based on decision tree.
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

# Here we are going to take the scenario of predicting the housing price based on couple of input data.
# So we will train the modal with some realistic house pricing data and then use the model to predict value of a house price.
area = [800, 1000, 1200, 1500, 1800, 2000, 2200, 2500]
rooms = [2, 2, 3, 3, 4, 4, 4, 5]
age = [20, 15, 10, 8, 5, 3, 2, 1]
costs = [30000, 40000, 45000, 46500, 55000, 75000, 90000, 150000]

data_frame = pd.DataFrame({
    'Area': area,
    'Rooms': rooms,
    'Age': age,
    'Cost': costs,
})

data_frame
```

```python
# Now create the data classification for training the model
X = data_frame[['Area', 'Rooms', 'Age']]
Y = data_frame[['Cost']]
```

```python
# Now we will create the model and train it
# max_depth is a hyperparameter that controls the maximum depth of the tree.
model = DecisionTreeRegressor(max_depth=3)
model.fit(X,Y)
```

```python
# Now model is trained to use and we can use it for doing the prediction and making decision
area_input = int(input("Enter Area: "))
rooms_input = int(input("Enter Rooms: "))
age_input = int(input("Enter Age: "))
predicted_cost = model.predict(pd.DataFrame({
    'Area': [area_input],
    'Rooms': [rooms_input],
    'Age': [age_input],
}))
print(f'Predicted Cost: {predicted_cost}')
```

```python
# Draw the decision tree graph
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(15, 10))
plot_tree(
    model,
    feature_names=['Area', 'Rooms', 'Age'],     # Names of your input features
    filled=True,                                # Colors nodes by their predicted value
    rounded=True,                               # Rounds the corners of the node boxes
    fontsize=10                                 # Adjust font size for legibility
)
plt.show()
```
