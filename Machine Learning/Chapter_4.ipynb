```python
# Training using Random Forest Algorithm
# In this case it will create multiple decision tree and use its average for calculating the better prediction

# Lets say we have scenario of dependency of marks obtained on study hours, sleep hours and practice tests
import pandas as pd
from sklearn.tree import DecisionTreeRegressor # We will use this for comparison
from sklearn.ensemble import RandomForestRegressor

study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
sleep_hours = [8, 7, 7, 6, 6, 5, 5, 4]
practice_tests_taken = [0, 1, 1, 2, 2, 3, 3, 4]
makrs_obtained = [30 ,40 ,50, 65, 70, 80, 85, 90]

data_frame = pd.DataFrame({
    'Study Hours': study_hours,
    'Sleep Hours': sleep_hours,
    'Practice Tests Taken': practice_tests_taken,
    'Marks Obtained': makrs_obtained,
})

data_frame
```

```python
X = data_frame[['Study Hours', 'Sleep Hours', 'Practice Tests Taken']]
Y = data_frame['Marks Obtained']
```

```python
decision_tree_model = DecisionTreeRegressor(max_depth=3)
decision_tree_model.fit(X, Y)
```

```python
random_forest_model = RandomForestRegressor(n_estimators=50, max_depth=3, random_state=42)
# n_estimators indicates how many decision tree need to be used for the model
# max_depth indicates the depth of each decision tree
# random_state indicates the randomness of the input values
random_forest_model.fit(X, Y)
```

```python
study_hour = int(input("Enter Study Hours: "))
sleep_hour = int(input("Enter Sleep Hours: "))
practice_test_taken = int(input("Enter number of practice test taken: "))
```

```python
input_data = pd.DataFrame({
    'Study Hours': [study_hour],
    'Sleep Hours': [sleep_hour],
    'Practice Tests Taken': [practice_test_taken]
})
```

```python
decision_tree_model_predict = decision_tree_model.predict(input_data)
random_forest_model_predict = random_forest_model.predict(input_data)
```

```python
print(f'Predicted Marks using Decision Tree: {decision_tree_model_predict}')
print(f'Predicted Marks using Random Forest: {random_forest_model_predict}')
```
