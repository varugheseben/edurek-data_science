```python
##### This program will train the modal using keras mnist dataset. So it will train model with multiple number images and then it predicts a 
# number which user picked from test data.
# Then from that data set it uses some for training and some for testing
# So we will have 
#       X_train and y_train for training 
#       X_test and y_test for testing
# X represents the input features (the data the model looks at), 
# while y represents the target labels (the answers the model is trying to guess). 
# We normalize inputs to help the math work better, but you do not normalize discrete category labels.
# So then we normalize the X inputs(X_train and X_test) by dividing with 255 so it will scale down between 0.0 and 1.0
# Then we create deep learning model using  models.Sequential(), where
#    layers.Flatten means Neural networks require standard flat arrays as inputs. This converts a 28x28 pixel grid into a single row of 784 numbers.
#    layers.Dense: This is the core engine of deep learning. Every input neuron connects to every output neuron in this layer, 
#                  calculating weighted sums to extract features.
#    optimizer='adam': The optimization algorithm. It acts like a guide, telling the network exactly how to tweak its internal weights to 
#                      reduce mistakes based on the errors it makes.
#    model.fit(): This kicks off the actual training loop where tensors flow through the graph, errors are calculated, and the model learns.
# This example is a classification type as the input values are not continuos

import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# 1. Load and prepare the MNIST dataset
# The dataset contains 60,000 training images and 10,000 testing images
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values to be between 0.0 and 1.0 (helps the network learn faster)
# This is for generating values in 0 and 1
X_train, X_test = X_train / 255.0, X_test / 255.0

# 2. Build the Deep Learning Model Architecture
# We use a Sequential model, which stacks layers linearly
model = models.Sequential([
    # Flatten the 2D image (28x28 pixels) into a 1D vector (784 pixels)
    layers.Flatten(input_shape=(28, 28)),
    
    # Dense (Fully Connected) hidden layer with 128 neurons and ReLU activation function
    layers.Dense(128, activation='relu'),
    
    # Dropout layer to prevent overfitting by randomly setting 20% of inputs to 0
    layers.Dropout(0.2),
    
    # Output layer with 10 neurons (one for each digit 0-9) using Softmax 
    # to output probability distributions
    layers.Dense(10, activation='softmax')
])

# 3. Compile the Model
# Define how the model measures performance (loss) and updates itself (optimizer)
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 4. Train the Model
# Feed the training data to the network for 5 iterations (epochs)
print("--- Training the Model ---")
model.fit(X_train, y_train, epochs=5)

# 5. Evaluate the Model
# Test the trained network on unseen data to check its true performance
print("\n--- Evaluating on Test Data ---")
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)

print(f"\nTest Accuracy: {test_acc * 100:.2f}%")

# 6. Make a Prediction
# Predict the digit
# We have 10000 test data, enter index number from that set
index_number = int(input("Enter number position: "))
prediction = model.predict(X_test[index_number:(index_number + 1)])
predicted_digit = np.argmax(prediction)

print(f"Predicted Digit: {predicted_digit}")
print(f"Actual Digit: {y_test[:index_number + 1]}")
```
