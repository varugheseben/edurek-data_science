```python
# Example of doing prediction using a neural network with one hidden layer
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense

X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

model = Sequential()
model.add(Dense(
    1,
    activation='sigmoid', # For categorical dependent variable we use sigmoid as the activation function
    input_shape=(1,) # Because our input has only one column
))

# Optimizer is sgd(Stochastic Gradient Descent), because we have continuous data in independent variable X
# Loss Function is 'binary_crossentropy' because we have two categorical data in dependent variable and activation is sigmoid
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=50, verbose=1)
for i in range(1,7):
  prediction = model.predict(np.array([[2]]))
  print(f"\nPredicted probability for {i}: {prediction[0][0]:.4f}")
```

```python
# In the above example prediction is not coming correctly because probabality is greater than 0.5 for numbers 1, 2,3 .
# This is due to less training data. So we can increase the training data by increasing the number of epochs
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense

X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

model = Sequential()
model.add(Dense(
    1,
    activation='sigmoid', # For categorical dependent variable we use sigmoid as the activation function
    input_shape=(1,) # Because our input has only one column
))

# Optimizer is sgd(Stochastic Gradient Descent), because we have continuous data in independent variable X
# Loss Function is 'binary_crossentropy' because we have two categorical data in dependent variable and activation is sigmoid
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=500, verbose=0)
for i in range(1, 7):
  prediction = model.predict(np.array([[i]]))
  print(f"\nPredicted probability for {i}: {prediction[0][0]:.4f}")
```

```python
# We still see its not predicting values correctly. For 3 its predicted value greater than 0.5
# So we will fix the randomness of weights generated in hidden layer and increase the epochs and use another optimizer algorithm
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense

keras.utils.set_random_seed(42) # -> This is for fixing the randomness for finding the weights probablity
X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

model = Sequential()
model.add(Dense(
    1,
    activation='sigmoid', # For categorical dependent variable we use sigmoid as the activation function
    input_shape=(1,) # Because our input has only one column
))

# Optimizer is Adam for better accuracy
# Loss Function is 'binary_crossentropy' because we have two categorical data in dependent variable and activation is sigmoid
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=500, verbose=0)
for i in range(1, 7):
  prediction = model.predict(np.array([[i]]))
  print(f"\nPredicted probability for {i}: {prediction[0][0]:.4f}")
```
