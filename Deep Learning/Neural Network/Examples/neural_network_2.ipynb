```python
# Example of using multi dimensional input data to a neural network and do the prediction
# So we have a data set of different students study pattern.
# So in input data we have hours studied and percentage of attendance
# And then we have dependent column specifying pass or fail

import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense

study_pattern = np.array([
    [1, 25],
    [2, 28],
    [3, 35],
    [4, 30],
    [4, 95],
    [5, 30],
    [5, 85],
    [3, 95],
    [5, 95],
    [2, 10]
])

result_pattern = np.array([0, 0, 0, 0, 1, 0, 1, 1, 1, 0])
prediction_model = Sequential()
prediction_model.add(Dense(
    1,
    activation ='sigmoid',
    input_shape = (2, )
))

prediction_model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
prediction_model.fit(study_pattern, result_pattern, epochs=100, verbose=0)
prediction = prediction_model.predict(np.array([[2, 90]]))

if (prediction[0] > 0.5):
  print('PASS')
else:
  print('FAIL')
```
