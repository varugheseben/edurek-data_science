```python
# Generate a 3x3 matrix of random numbers using NumPy.

import numpy as np

# Generate a 3x3 matrix of random floats between 0 and 1
print("3x3 matrix of random floats between 0 and 1")

matrix = np.random.rand(3,3)
print(matrix)

# 3x3 matrix with integers between 1 and 10
print("\n\n3x3 matrix with integers between 1 and 10")
int_matrix = np.random.randint(1, 11, size=(3,3))
print(int_matrix)

# 3x3 matrix from a standard normal distribution
print("\n\n3x3 matrix from a standard normal distribution")
normal_matrix = np.random.randn(3,3)
print(normal_matrix)
```
