```python
# There are different arithmetic operations can be performed using Numpy
# 1) Scalar Arithmetic
# 2) Arithmetic using vector math functions
# 3) Element-wise arithmetic
# 4) Comparison arithmetic
```

```python
# Scalar Arithmetic
import numpy as np

array = np.array([1, 2, 3, 4])
print(array + 1)
print(array * 2)
print(array - 1)
print(array / 2)
print(array ** 2)
```

```python
# Arithmetic using vector math functions

import numpy as np

array = np.array([4, 9, 16, 25])
print(np.sqrt(array))

# Examples of mathematical constants defined in numpy
print(np.pi)
```

```python
# Element-wise arithmetic

import numpy as np

array_1 = np.array([2, 3, 5 ,7])
array_2 = np.array([10, 11, 12, 13])

print(array_1 + array_2)
print(array_1 - array_2)
print(array_1 * array_2)
print(array_1 / array_2)
print(array_1 ** array_2)
```

```python
# Comparison Arithmetic
# With this arithmetic we can create boolean arrays

import numpy as np
marks = np.array([95, 65, 58, 85, 74, 100])
print(marks > 65)
```

```python
print(marks <= 65)
```

```python
marks = np.array([95, 65, 58, 85, 74, 100])
marks[marks < 65] = 0
print(marks)
```

```python
# Similarly
marks = np.array([95, 65, 58, 85, 74, 100])
marks[marks > 65] = 1
print(marks)
```
