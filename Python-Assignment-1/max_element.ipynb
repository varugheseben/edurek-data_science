```python
# Write a function that takes a list and returns the maximum element.

def get_max(elements):
	return max(elements)

number = int(input("Number of elements in list: "))

elements = []

while number > 0:
    try:
        element = int(input("Enter list element: "))
        elements.append(element)
        number = number - 1
    except Exception as e:
        print("Invalid element, please enter a number")

print(f'Your list: {elements}')
print(f'Max value is: {get_max(elements)}')
```
