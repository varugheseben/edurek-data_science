```python
#Decision tree
area = [800, 1000, 1200, 1500, 1800, 2000, 2200, 2500]
rooms = [2, 2, 3, 3, 4, 4, 4, 5]
age = [20, 15, 10, 8, 5, 3, 2, 1]
#pricing
price = [30, 40, 50, 65, 80, 90, 100, 120]
import pandas as pd
df = pd.DataFrame({
    "Area": area,
    "Rooms": rooms,
    "Age": age,
    "Price": price
})
df
```

```python
from sklearn.tree import DecisionTreeRegressor
X = df[["Area", "Rooms", "Age"]]
Y = df["Price"]
X
# model = DecisionTreeRegressor
```

```python
Y
```

```python
model = DecisionTreeRegressor(max_depth=3)
model.fit(X,Y)
```

```python
prediction = model.predict([[1200, 5, 1]])
print("Predicted value:", prediction)
```

```python
## Random forest regressor
# predict student marks based on :  sleep hours, study hours, number of practice test

study_hours = [1,2,3,4,5,6,7,8]
sleep_hours = [8,7,7,6,6,5,5,4]
practice_tests = [0,1,1,2,2,3,3,4]

marks = [30,40,50,70,60,80,85,90]

import pandas as pd
df = pd.DataFrame({ 
    "StudyHours": study_hours,
    "SleepHours": sleep_hours,
    "PracticeTests": practice_tests,
    "Marks": marks
})
df
```

```python
# #inference: 
# More study hours -> more marks
# More practice test -> better performance
# Too much sleep -> maybe less study
# model will learn combined effect 
```

```python
from sklearn.tree import DecisionTreeRegressor
X = df[["StudyHours", "SleepHours", "PracticeTests"]]
y = df["Marks"]
tree_model = DecisionTreeRegressor(max_depth=3)
tree_model.fit(X,y)
```

```python
# Random forest 

from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=50,
    max_depth=3,
    random_state=42
)
rf_model.fit(X,y)
    
```

```python
# prediction 
# example student:
# study = 5 hours , sleep = 6 hours , practice = 2 test

tree_pred = tree_model.predict([[5,6,2]])
rf_pred = rf_model.predict([[5,6,2]])

print("Decision Tree: ", tree_pred)
print("Random Forest: ", rf_pred)
```
